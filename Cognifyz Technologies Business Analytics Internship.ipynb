# Data Overview

### Objective: Understand the dataset structure.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("cogdata.csv")

# Descriptive Statistics
df_info = df.info()

# Display the dataset info (number of entries, columns, and data types)
print(df_info)
```

The dataset contains 40 entries and 24 columns. Here's a summary of the columns and their data types:

- **8 columns** are of type `int64` (numerical):
  - `age`, `Mutual_Funds`, `Equity_Market`, `Debentures`, `Government_Bonds`, `Fixed_Deposits`, `PPF`, `Gold`
  
- **16 columns** are of type `object` (categorical or textual):
  - `gender`, `Investment_Avenues`, `Stock_Marktet`, `Factor`, `Objective`, `Purpose`, `Duration`, `Invest_Monitor`, `Expect`, `Avenue`, `What are your savings objectives?`, `Reason_Equity`, `Reason_Mutual`, `Reason_Bonds`, `Reason_FD`, `Source`
  
All columns have non-null values, meaning there are no missing values in the dataset.

# Gender Distribution

### Objective: Visualize gender distribution in the dataset.

```python
import matplotlib.pyplot as plt

# Extract Gender Information
gender_counts = df['gender'].value_counts()

# Visualization (Bar Chart)
plt.figure(figsize=(6, 4))
gender_counts.plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# Visualization (Pie Chart)
plt.figure(figsize=(6, 6))
gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen'])
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

The bar chart clearly highlights the frequency of each gender, while the pie chart provides a more intuitive understanding of the proportion of each gender in the dataset. These visualizations make it easier to identify any gender imbalance among the participants.

# Reasons for Investment

### Objective: Analyze and summarize reasons for investment choices.

```python
# Explore Reasons Columns
reasons_columns = ['Reason_Equity', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD']

# Checking the unique reasons given for each investment avenue
reasons_summary = {}
for column in reasons_columns:
    reasons_summary[column] = df[column].value_counts()

# Display the summary of reasons for each avenue
reasons_summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# Combine all reason columns into a single DataFrame for ease of visualization
reasons_columns = ['Reason_Equity', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD']
reasons_data = df[reasons_columns]

# Melt the data to have a single column of reasons, categorized by investment type
melted_reasons = reasons_data.melt(var_name='Investment_Type', value_name='Reason')

# Plot the frequency of reasons for each investment type
plt.figure(figsize=(12, 8))
sns.countplot(y='Reason', hue='Investment_Type', data=melted_reasons, order=melted_reasons['Reason'].value_counts().index)

# Add title and labels
plt.title('Frequency of Reasons for Different Investment Types', fontsize=16)
plt.xlabel('Count')
plt.ylabel('Reason')

# Show the plot
plt.tight_layout()
plt.show()
```

1. Capital Appreciation and Dividends in Equity:
The most common reasons for investing in equities are "Capital Appreciation" and "Dividend."
This suggests that participants are attracted to equities for the potential long-term growth (capital appreciation) and regular income (dividends).

2. Better Returns and Fund Diversification in Mutual Funds:
Mutual fund investors focus on "Better Returns" and "Fund Diversification."
It indicates that participants view mutual funds as a way to potentially achieve higher returns while managing risk by diversifying their investments.

3. Safety and Assured Returns in Bonds:
"Safe Investment" and "Assured Returns" are the most frequent reasons for investing in bonds.
Investors view bonds as a stable, low-risk option for securing their capital and ensuring consistent returns, likely appealing to risk-averse individuals.

4. Risk-Free and Fixed Returns in Fixed Deposits (FDs):
For fixed deposits, "Risk-Free" and "Fixed Returns" are dominant reasons.
This shows that participants seek safety and guaranteed returns, often prioritizing security over higher potential yields, typically a preference among conservative investors.

5. Themes Across Investment Types:
Risk-Return Tradeoff: Investors balance the desire for better returns (equities and mutual funds) with safety and stability (bonds and fixed deposits).
Tax Benefits: There are also mentions of "Tax Benefits" in equities and bonds, highlighting the role tax advantages play in investment decisions.
These insights show diverse investment goals: growth and income generation on one side, and safety with guaranteed returns on the other. Investors seem to select avenues based on their risk appetite, time horizon, and financial objectives.
